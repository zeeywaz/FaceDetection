Importing Libraries

```python
import cv2
import face_recognition
```

Loading and Encoding Reference Images

```python
# Load the reference image
img = cv2.imread("messi.jpg")
# Convert the image from BGR (OpenCV default) to RGB
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Generate the face encoding for the reference image
img_encoding = face_recognition.face_encodings(rgb_img)[0]
```

Loading and Encoding the Target Image

```python
# Load the target image
img2 = cv2.imread("images/mark.jpg")
# Convert the target image to RGB
rgb_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
# Generate the face encoding for the target image
img_encoding2 = face_recognition.face_encodings(rgb_img2)[0]
```

Comparing the Encoded Faces

```python
# Compare the two face encodings
result = face_recognition.compare_faces([img_encoding], img_encoding2)
print("Result: ", result)
```

Setting Up Real-Time Face Recognition with Webcam

```python

from simple_facerec import SimpleFacerec

# Create an instance of SimpleFacerec
sfr = SimpleFacerec()
# Load and encode all images from the 'images/' directory
# Load Camera
cap = cv2.VideoCapture(0)

sfr.load_encoding_images("images/")
while True:
    ret, frame = cap.read()
    # Detect known faces in the current frame
    face_locations, face_names = sfr.detect_known_faces(frame)
    for face_loc, name in zip(face_locations, face_names):
        y1, x2, y2, x1 = face_loc[0], face_loc[1], face_loc[2], face_loc[3]
        # rectangle around the detected face
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 200), 4)
        # Display the name of the recognized person
        cv2.putText(frame, name, (x1, y1 - 10), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 200), 2)
    cv2.imshow("Frame", frame)
    key = cv2.waitKey(1)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
